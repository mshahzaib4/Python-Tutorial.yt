# **Relationship plot**

```python
# import Library
import matplotlib.pyplot as plt
```

```python

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

plt.show()
```

```python
# Draw line plot

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')

plt.show()
```

```python
# Draw line plot

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.hexbin(x, y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Hexbin Plot')

plt.show()
```

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Dataset/Dataset.csv')

# Create a figure with subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Subplot 1: Scatter plot
axes[0].scatter(data['Name'], data['Age'])
axes[0].set_xlabel('Name')
axes[0].set_ylabel('Age')
axes[0].set_title('Scatter Plot of Age by Name')

# Subplot 2: Line plot
axes[1].plot(data['Name'], data['Age'], color='red')
axes[1].set_xlabel('Name')
axes[1].set_ylabel('Age')
axes[1].set_title('Line Plot of Age by Name')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
