# **Modules and Packages**


### **Understanding Modules and Packages in Python**

#### **Modules**

A module is a Python file containing code like functions, classes, and variables. Modules help break code into smaller, more manageable pieces, making it easier to understand.

**Advantages of Modules**
- **Modularization** Breaks code into smaller, reusable components.
- **Organization** Helps maintain a structured codebase.
- **Reuse** Saves time by allowing the reuse of code across different parts of a program.

#### **Creating and Importing Modules**

## **`1.` Creating and Importing a Module**

Create a Python file named `name.py` with the following code

```python
# name.py

def get_name():
    return "Muhammad Shahzaib Yaqoob"
```
Create another Python file named `name2.py` with the following code

```python
# name.py
import name

print(name.get_name())
```

## **`2.` Exploring a Standard Library Module**

**Importing the `math` module**
```python
import math

# Use the pi constant
print(f"Value of pi: {math.pi}")

# Use the sqrt function
result = math.sqrt(16)
print(f"Square root of 16: {result}")
```

#### **Packages**
Packages are groups of modules that help organize code in a hierarchical structure, improving organization and enabling code reuse.

## **`3.` Installing and Using an External Package**

**Installing numpy package using pip**
```bash
pip install numpy
```

**Usage in Python**
```python
import numpy as np

# Create a numpy array
arr = np.array([1, 2, 3, 4])

# Print the array
print(arr)
```
