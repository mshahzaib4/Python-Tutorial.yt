```python
import matplotlib.pyplot as plt
```

```python

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.bar(x, y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Plot')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Dataset/Dataset.csv')

plt.pie(data["Age"], labels=data["Age"], autopct='%1.1f%%')
plt.title('Age Distribution')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
axes[0].pie(data["Age"], labels=data["Age"], autopct='%1.1f%%')
axes[0].set_title('Age Distribution')

axes[1].bar(data["Name"], data["Age"])
axes[1].set_title('Name vs Age')

plt.show()
```
