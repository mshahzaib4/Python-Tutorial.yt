```python
import matplotlib.pyplot as plt
import pandas as pd

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.hist(x, y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram Plot')

plt.show()
```

```python
data=pd.read_csv("Dataset\Dataset.csv")

plt.boxplot(data["Age"], patch_artist=True, boxprops=dict(facecolor='red'))

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Boxplot of Age')

# Show the plot
plt.show()
```
